```python
import os
import pandas as pd
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns
```

Importazione dei dataset e dichiarazione di variabili utili

```python
data = pd.read_pickle(os.path.join(os.path.join("", "data"), "data.pickle"))

graphics_folder_path = os.path.join('', 'graphics')

correlation_gdp_file_path = os.path.join(graphics_folder_path, "correlation_gdp.png")
correlation_life_expectancy_file_path = os.path.join(graphics_folder_path, "correlation_Life expectancy.png")
correlation_unemployment_file_path = os.path.join(graphics_folder_path, "correlation_unemployment.png")
correlation_homicide_rate_file_path = os.path.join(graphics_folder_path, "Homicide rate.png")
correlation_literacy_file_path = os.path.join(graphics_folder_path, "correlation_literacy.png")
correlation_growth_rate_file_path = os.path.join(graphics_folder_path, "correlation_growth_rate.png")
correlation_population_file_path = os.path.join(graphics_folder_path, "correlation_population.png")
correlation_density_file_path = os.path.join(graphics_folder_path, "correlation_density.png")
correlation_file_path = os.path.join(graphics_folder_path, "correlation.png")
```

Come sono correlati la felicit√† e il PIL?

```python
corr_gdp = sp.stats.spearmanr(data["Happiness"], data["GDP per capita"])

fig, ax = plt.subplots(figsize = (10, 5))
ax = sns.regplot(data=data, x="Happiness", y="GDP per capita", line_kws=dict(color="#D4AF37"), )
plt.xlabel("Happiness")
plt.ylabel("GDP per capita")
ax.set_ylim(bottom=-1500)
ax.annotate(f"Spearman={corr_gdp[0]:.4f} with p={corr_gdp[1]:.2e}", xy=(0.02, 0.96), xycoords='axes fraction', ha='left', va='center', bbox={'boxstyle': 'round', 'fc': 'white', 'ec': 'lightgrey'})
plt.savefig(correlation_gdp_file_path, bbox_inches='tight')
plt.show()
```
